```python
## Create a list for store each individual best estimator
best_estimators = []
```

```python
## parameters
params = {
    "learning_rate": [.1, .5],
    "colsample_bytree": [.3,.5,],
    "max_depth": [2, 4],
    "alpha": [3, 5],
    "subsample": [.5],
    "n_estimators": [30, 70],
    "random_state" : [42]
}

## XGBoost Regressor
XGBR =  XGBRegressor()
XGBR_grid = GridSearchCV(XGBR, params, scoring='r2', cv=7, n_jobs=-1)

XGBR_grid.fit(x_train, y_train)

## Output
print("Best parameters:  {}:".format(XGBR_grid.best_params_))
print("Best score: {}".format(XGBR_grid.best_score_))

## Append to list
best_estimators.append(["XGBoostR",XGBR_grid.best_estimator_])
```

```python
## parameters
params = {
    "max_depth": ["None",10],
    "max_features": ["auto",.3, .4],
    "min_samples_leaf": [3,7],
    "min_samples_split": [2, 4],
    "n_estimators": [50, 100],
    "random_state" : [42]
}

## ExtraTreesRegressor
EXTR =  ExtraTreesRegressor()
EXTR_grid = GridSearchCV(EXTR, params, scoring='r2', cv=7, n_jobs=-1)

EXTR_grid.fit(x_train, y_train)

## Output
print("Best parameters:  {}:".format(EXTR_grid.best_params_))
print("Best score: {}".format(EXTR_grid.best_score_))

## Append to list
best_estimators.append(["ExtraTreesR",EXTR_grid.best_estimator_])
```

```python
## Parameters
params = {
    "max_depth": [2, 3],
    "max_features": ["auto",0.3],
    "min_samples_leaf": [1,3],
    "min_samples_split": [2, 5],
    "n_estimators": [50, 100],
    "random_state" : [42],
    "tol" : [0.0001,0.01]
}

## GradientBoostingRegressor
GBR = GradientBoostingRegressor()
GBR_grid = GridSearchCV(GBR, params, scoring='r2', cv=7, n_jobs=-1)
GBR_grid.fit(x_train, y_train)

## Output
print("Best parameters:  {}:".format(GBR_grid.best_params_))
print("Best score: {}".format(GBR_grid.best_score_))

## Append to list
best_estimators.append(["GradientBoostR",GBR_grid.best_estimator_])
```

# Evaluating optimized Estimators

```python
from sklearn.model_selection import cross_val_score,KFold

#storing cross_val_score estimators into empty dataframe
#note: "cv" for mean cv & "cv_all" for later visualization use
evaluate = pd.DataFrame(columns=["model","std","cv_mean","cv_median"])

for name,estimator in best_estimators:
    kfold=KFold(n_splits=10,random_state=42,shuffle=True)
    cv = cross_val_score(estimator,x_train,y_train,cv=kfold,scoring='r2',n_jobs=-1)
    
    row=evaluate.shape[0]
    evaluate.loc[row,'model']=name
    evaluate.loc[row,'cv_mean']=round(cv.mean(),3)
    evaluate.loc[row,'cv_median'] = round(np.median(cv),3)
    evaluate.loc[row,'std'] = '+/- {}'.format(round(cv.std(),4))
    
    evaluate = evaluate.sort_values("cv_mean",ascending=False)
    
    
evaluate
```

```python
# Setting up the figure size
fig, ax = plt.subplots(figsize=(12, 9))

# Creating the bar chart with evaluate["model"] on the x-axis and evaluate["cv_mean"] on the y-axis
bar = sns.barplot(x=evaluate["model"], y=evaluate["cv_mean"], palette=sns.color_palette("rocket"), ax=ax)

# Adding text labels to each bar with the height value
for rec in bar.patches:
    height = rec.get_height()
    ax.text(rec.get_x() + rec.get_width() / 2, height * 1.02, height, ha="center")

# Displaying the bar chart
plt.show()
```
